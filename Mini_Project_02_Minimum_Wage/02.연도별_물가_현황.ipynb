```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 불러오기
price = pd.read_excel('./data/연도별_물가지수.xls', header=2, skipfooter=3)
price
```

```python
# column명 재정의
price = price.rename(columns={'Unnamed: 0' : '품목'})
price
```

```python
# 데이터 전처리(결측치, 이상치)
price.info()
price.describe()
```

```python
# 데이터 전처리(중복치)
price[price.duplicated()]
```

```python
# 데이터 정제(필요 column 추출)
price =price[['품목', '2016', '2017', '2018', '2019', '2020', '2021']]
price
```

```python
# 소비자 물가 총 지수 추출

price_all = price[0:1].T.reset_index()
price_all = price_all[1:]
price_all.columns = ['연도','물가지수']
price_all = price_all.reset_index(drop =True)
price_all
```

```python
# 소비자 물가 상승률 추출
price_all_percent = price[1:2].T.reset_index()
price_all_percent = price_all_percent[1:]
price_all_percent.columns = ['연도','물가상승률(%)']
price_all_percent = price_all_percent.reset_index(drop =True)
price_all_percent
```

```python
# 품목별 물가 상승률 추출
price_item = price[2:].T.reset_index()
price_item = price_item[1:]
price_item.columns =['연도','식료품/비주류','주류/담배','의류/신발','주택/수도/전기/연료',
                      '가정용품/가사서비스','보건','교통','통신','오락/문화','교육','음식/숙박','기타상품/서비스']
```

```python
# 정제된 데이터 시각화(물가 상승률)
ax1_ylist = np.arange(0,180,30)
ax2_ylist = np.arange(0,5.5,0.5)

plt.rc('font', family = 'Malgun Gothic', size = 18)
fig, ax1 = plt.subplots(figsize=(20, 10))
fig.set_facecolor('#EEEEEE')

ax1.set_title('연도별 물가지수 및 증가율',fontsize = 20,loc='center', pad=30, fontweight="bold")
ax1.text(-0.48,151.5, '단위: %', fontsize=15, color = 'black')
plt.xticks([0,1,2,3,4,5], price_all_percent['연도']+'년')

ax1.bar(price_all['연도'],price_all['물가지수'], width=0.4 ,color ='navy')
ax1.set_facecolor('#EEEEEE')
ax1.set_yticks(ax1_ylist)

# 그래프 상단에 값 추가하기
for i in range(len(price_all['물가지수'])):
    height = price_all['물가지수'][i:i+1]
    plt.text(price_all['연도'][i:i+1], height + 0.12, '%.1f' %height, ha='center', va='bottom', size = 15)

ax2 = ax1.twinx()
ax2.plot(price_all_percent['연도'],price_all_percent['물가상승률(%)'], marker='o', ms= 15,color ='#F2B117')

plt.rcParams['lines.linewidth'] = 4

ax2.set_yticks(ax2_ylist)
ax2.text(5.2,5.05, '단위: %', fontsize=15, color = 'black')

#그래프 상단에 값 추가하기
for i in range(len(price_all_percent['물가상승률(%)'])):
    height = price_all_percent['물가상승률(%)'][i:i+1]
    plt.text(price_all_percent['연도'][i:i+1], height + 0.15, '%.1f' %height, ha='center', va='bottom', size = 15 , color='white', fontweight="bold")

plt.show()
```
