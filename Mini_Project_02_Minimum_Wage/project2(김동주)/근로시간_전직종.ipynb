```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
file_path = "./data/직종별_임금_및_근로시간.xlsx"
data_time = pd.read_excel(file_path)
data_time
```

```python
data_time = data_time.transpose()
data_time
```

```python
data_time = data_time.iloc[1:,:]
data_time = data_time.reset_index()
data_time
```

```python
data_time.columns = ["연도", '관리자', '전문가 및 관련 종사자',	'사무 종사자', '서비스 종사자',	'판매 종사자', '농림어업 숙련 종사자', '기능원 및 관련 기능 종사자', '장치,기계조작 및 조립종사자',	'단순노무 종사자']
data_time
```

```python
data_time = data_time.set_index("연도")
data_time
```

```python

data_time.plot(title = 'SomeTitle', figsize=(20,10), legend = True, marker = 'o', rot=0)


plt.xlabel('연도')
plt.ylabel('근무시간')
plt.title('연도별 근무시간')

plt.show()
```

```python
color = ['#F2B117','#F2B117','#F2B117','navy','navy','#F2B117','#F2B117','#F2B117','navy']
data_time.plot(title = 'SomeTitle', figsize=(20,10), legend = True, marker = 'o', rot=0, color = color)


plt.xlabel('연도')
plt.ylabel('근무시간')
plt.title('직종별 근무시간')

plt.show()
```
