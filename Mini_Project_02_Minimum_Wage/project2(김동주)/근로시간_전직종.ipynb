```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
file_path = "./data/직종별_임금_및_근로시간.xlsx"
data_time = pd.read_excel(file_path)
data_time
```

```python
data_time = data_time.transpose()
data_time
```

```python
data_time = data_time.iloc[1:,:]
data_time = data_time.reset_index()
data_time
```

```python
data_time.columns = ["연도", '관리자', '전문가 및 관련 종사자',	'사무 종사자', '서비스 종사자',	'판매 종사자', '농림어업 숙련 종사자', '기능원 및 관련 기능 종사자', '장치,기계조작 및 조립종사자',	'단순노무 종사자']
data_time
```

```python
data_time.columns[1:]
```

```python
color = ['navy','navy','navy','#F2B117','navy','navy','navy','navy','navy']
ylist = np.arange(130,210,10)

fig = plt.figure(figsize=(20,10))

plt.title('연도 및 직종별 월근무시간 현황', fontsize = 20,loc='center', pad=30, fontweight="bold")

fig.set_facecolor('#EEEEEE')
plt.gca().set_facecolor('#EEEEEE')
plt.rc('font', family = 'Malgun Gothic', size = 18)

for i in range(len(data_time.columns[1:])):
    plt.plot(data_time['연도'],data_time[data_time.columns[i+1]],marker = 'o', ms= 8, color = color[i])


for i in range(len(data_time['연도'])):
    height = data_time['서비스 종사자'][i:i+1]
    plt.text(data_time['연도'][i:i+1], height + 0.8, '%.1f' %height, ha='center', va='bottom', size = 15, 
             color='black')
    
plt.rcParams['lines.linewidth'] = 2

plt.text(-0.38,201, '단위: 시간', fontsize=15, color = 'black')   

plt.yticks(ylist)
plt.legend(data_time.columns[1:], fontsize=10)
plt.show()
```
