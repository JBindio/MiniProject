```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
file_path = "./산업_규모별_고용_20220721112015.xlsx"
data = pd.read_excel(file_path)
data
```

```python
# 년도별 데이터 분할

dt_2018 = data.iloc[0:,:8]
dt_2019 = data.iloc[0:,8:16]
dt_2020 = data.iloc[0:,16:24]
dt_2021 = data.iloc[0:,24:32]
```

```python
# 열 이름 변경
dt_2018.columns = ["산업분류","종사자_전체","근로자_상용","근로자_임시일용","종사자_기타","채용_전체","채용_상용","채용_임시일용"]
dt_2019.columns = ["산업분류","종사자_전체","근로자_상용","근로자_임시일용","종사자_기타","채용_전체","채용_상용","채용_임시일용"]
dt_2020.columns = ["산업분류","종사자_전체","근로자_상용","근로자_임시일용","종사자_기타","채용_전체","채용_상용","채용_임시일용"]
dt_2021.columns = ["산업분류","종사자_전체","근로자_상용","근로자_임시일용","종사자_기타","채용_전체","채용_상용","채용_임시일용"]
```

```python

dt_18 = dt_2018[["산업분류","종사자_전체"]]
dt_19 = dt_2019[["산업분류","종사자_전체"]]
dt_20 = dt_2020[["산업분류","종사자_전체"]]
dt_21 = dt_2021[["산업분류","종사자_전체"]]
```

```python
# 조인, 열이름 변경
dt_1 = pd.merge(dt_18,dt_19,how="inner",on= "산업분류")
dt_2 = pd.merge(dt_1,dt_20,how="inner",on= "산업분류")
dt_3 = pd.merge(dt_2,dt_21,how="inner",on= "산업분류")
dt_3
dt_3.columns = ["산업분류","18_종사자_전체","19_종사자_전체","20_종사자_전체","21_종사자_전체"]
dt_3
```

```python
# 전체와 나머지 분리
dt_all = dt_3[dt_3['산업분류'] == '전체']
dt_no_all = dt_3[dt_3['산업분류'] != '전체'].reset_index(drop=True)
dt_no_all
```

```python
#산업분류에 맨앞 영어만 가져옴
dt_8 =[]
for i in range(0,18):
    dt_8.append(dt_no_all["산업분류"][i][0])

dt_no_all["산업분류"] = dt_8
dt_no_all.columns = ["산업분류", "2018", "2019", "2020", "2021"]
dt_no_all
```

```python
dt_all.columns = ["산업분류", "2018", "2019", "2020", "2021"]
dt_all
```

```python
# 년도별 차이를 비율로 구함
dt_18_19 = dt_no_all["2019"] / dt_no_all["2018"]-1
```

```python
dt_19_20 = dt_no_all["2020"] / dt_no_all["2019"]-1
dt_19_20
```

```python
dt_20_21 = dt_no_all["2021"] / dt_no_all["2020"]-1
dt_20_21
```

```python
dt_18_21 = dt_no_all["2021"] / dt_no_all["2018"]-1
dt_18_21
```

```python
# 열붙이기
dt_no_all["산업분류"]
dt_f = pd.concat([dt_no_all["산업분류"],dt_18_19,dt_19_20,dt_20_21,dt_18_21], axis=1)
dt_f.columns = ["산업분류", "18-19", "19-20", "20-21", "18-21"]
dt_f
```

```python
ylist = np.arange(-0.2,0.3,0.1)
plt.figure(figsize=(20,10))
plt.rc('font', family = 'Malgun Gothic', size = 18)
plt.title('18-19차이비율',fontsize = 20,loc='center', pad=30, fontweight="bold")
ax = plt.bar(dt_f['산업분류'],dt_f['18-19'],width=0.4, color ='#F2B117')
plt.yticks(ylist)

plt.show()
```
